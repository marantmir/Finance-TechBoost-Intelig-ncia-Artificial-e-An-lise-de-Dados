# Estruturas de controle

Vejamos agora as estruturas de controle simples de Python.

O primeiro ponto a notar  que os blocos de comandos em Python so delimitados pela quantidade de indentao no texto do cdigo. Comandos consecutivos que so indentados pelo mesmo nmero de espaos em branco esto dentro do mesmo bloco. Um bloco  comeado pela presena de `:` no final da linha.

Por exemplo, o cdigo abaixo imprime "sei matemtica" se 2 for menor que 3.

```python
# > = maior , < = menor, == = igual, != = diferente
```

```python
2 < 3, 2 > 3
```

```python
if 2 < 3: #se
  print('Sei matemtica')
```

```python
if 2 > 3: #se
  print('Sei matemtica')
```

A expresso da condio no `if` no precisa de parntesis e pode ser to complexa quanto necessrio.

```python
#and = e
#or = ou
if 1 > 0 and -1 < 0: #if ((1 > 0) and (-1 < 0)):
    print('OK')
```

```python
#and = e
#or = ou
if 1 < 0 and -1 < 0: #if ((1 > 0) and (-1 < 0)):
    print('OK')
```

```python
#and = e
#or = ou
if 1 < 0 or -1 < 0: #if ((1 > 0) and (-1 < 0)):
    print('OK')
```

Quando queremos executar uma operao tambm quando a condio  falsa, usamos um `else`, com notao similar.

```python
a = 1
b = 2
if a > b: #se
    print('a  maior')
else: #seno
    print('a no  maior')
```

```python
a = 4
b = 2
if a < b: #se
    print('a  menor')
else: #seno
    print('a no  menor')
```

Se tivermos vrias condies a testar, podemos usar tantos `elif` (seguidos de condio) quanto necessrios. A condio de um `elif` somente ser testada se todas as condies anteriores forem falsas.

```python
a = 1
b = 1
if a < b: #se
    print('a  menor')
elif a == b: #seno se
    print('so iguais')
else: #seno
    print('a  maior')
```

*Loops* so expressos por `while`, seguido da condio de repetio.

```python
a = 0
while a <= 10: #enquanto
    print(a)
    a += 1 #a = a+1
print('Acabou')
```

#for = para
#in = dentro

Uma outra forma de executar *loops*  o `for`. Ele permite avaliar todos os valores de uma sequncia de valores fornecida.  A sintaxe :

    for x in seq:
        comandos

onde `seq`  uma "sequncia" de objetos (veremos em aulas posteriores a definio mais precisa), `x`  uma varivel que vai referenciar cada um dos objetos na ordem determinada pela sequncia, e `comandos` so os comandos a executar (em geral, fazem referncia a `x`).

```python
minhalista = [0, 2, 4, 6, 1, 3, 5, 7]
for x in minhalista: # para cada x em minhalista
    print(x/2)
```

- Para numeros pares -> resto da divisao sempre  0
- Para nmeros impares nao!

```python
minhalista = [0, 2, 4, 6, 1, 3, 5, 7]
for x in minhalista:
    if x % 2 != 0: # x % 2  um teste pra ver se o valor  par. % == resto da diviso
        print(x)
```

```python
minhalista = [0, 2, 4, 6, 1, 3, 5, 7]
for x in minhalista:
    if x % 2 == 0: # x % 2  um teste pra ver se o valor  par. % == resto da diviso
        print(x)
```
